```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(r'/content/sammy_face.jpg')
plt.imshow(img)
```

```python
img.size
```

```python
edges = cv2.Canny(image = img,threshold1 = 127,threshold2 = 127)#2 threshold values to filter out false edges
plt.imshow(edges)
```

```python
edges = cv2.Canny(image = img,threshold1 = 0,threshold2 = 255)#2 threshold values to filter out false edges
plt.imshow(edges)
```

```python
#to get efficient values of threshopl by help of median
med = np.median(img)
med
```

```python
low_thresh = int(max(0,0.7*med))
high_thresh = int(min(255,1.3*med))
```

```python
edges = cv2.Canny(image = img,threshold1 = low_thresh,threshold2 = high_thresh+100)#2 threshold values to filter out false edges
plt.imshow(edges)
```

```python
blurr_img = cv2.blur(img,ksize = (5,5))
```

```python
edges = cv2.Canny(image = blurr_img,threshold1 = low_thresh,threshold2 = high_thresh+50)#2 threshold values to filter out false edges
plt.imshow(edges)
```
